```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('twogaussians42.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
y_kmeans = kmeans.fit_predict(X)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_kmeans)
print("Accuracy score: {:.2f}".format(accuracy))
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 100, facecolors='green', edgecolors='green', label = 'Original 1', alpha = 1)
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 100, facecolors='black', edgecolors='gray', label = 'Original 2', alpha = 1)

plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], facecolors='none', edgecolors='green', label='Cluster 1',marker='*', alpha = 1)
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], facecolors='none', edgecolors='black', label='Cluster 2',marker='.', alpha = 1)


plt.title('K_Means')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
em = GaussianMixture(n_components=2, random_state=0)
y_kmeans = em.fit_predict(X)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_kmeans)
print("Accuracy score: {:.2f}".format(accuracy))
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 100, facecolors='green', edgecolors='green', label = 'Original 1', alpha = 1)
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 100, facecolors='black', edgecolors='gray', label = 'Original 2', alpha = 1)

plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], facecolors='none', edgecolors='green', label='Cluster 1',marker='*', alpha = 1)
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], facecolors='none', edgecolors='black', label='Cluster 2',marker='.', alpha = 1)


plt.title('GMM')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import SpectralClustering
sc = SpectralClustering(n_clusters=2, affinity='rbf', gamma=10)
y_kmeans = sc.fit_predict(X)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_kmeans)
print("Accuracy score: {:.2f}".format(accuracy))
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 100, facecolors='green', edgecolors='green', label = 'Original 1', alpha = 1)
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 100, facecolors='black', edgecolors='gray', label = 'Original 2', alpha = 1)

plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], facecolors='none', edgecolors='green', label='Cluster 1',marker='*', alpha = 1)
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], facecolors='none', edgecolors='black', label='Cluster 2',marker='.', alpha = 1)


plt.title('Spectral_Clustering')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method K_Means')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
