```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Load the dataset
data = pd.read_csv('twogaussians42.csv')

# separate features and labels
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python

# define the number of clusters
k_range = range(2, 6)

# create a figure with two rows and two columns
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# define colors and shapes for different labels
colors = ['red', 'blue']
shapes = ['*', '.', '^', 'd']

# loop over different k values
for i, k in enumerate(k_range):

    # create k-means model
    kmeans = KMeans(n_clusters=k)
    y_pred = kmeans.fit_predict(X)

    # plot decision boundary and clusters
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.1), np.arange(x2_min, x2_max, 0.1))
    Z = kmeans.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    axs[i//2, i%2].contourf(xx1, xx2, Z, alpha=0.4)
    for j in range(k):
        axs[i//2, i%2].scatter(X[y_pred == j, 0], X[y_pred == j, 1], c=colors[j%2], label='Label {}'.format(j+1), marker=shapes[j])
    axs[i//2, i%2].set_title("k = {}".format(k))
    axs[i//2, i%2].legend()

# adjust plot layout
plt.tight_layout()

# save plot
plt.savefig("kmeans_clusters.png")
```
