```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Load the dataset
dataset = pd.read_csv('twogaussians42.csv')

# separate features and labels
data = dataset.iloc[:, :-1].values
labels = dataset.iloc[:, -1].values
```

```python
# run k-means clustering for different values of k
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
fig.suptitle(' kmeans_clusters_twogaussians42 ')
axs = axs.ravel()
for i, k in enumerate(range(2, 6)):
    
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    preds = kmeans.predict(data)

    # plot the original classes and the k-means clusters
    ax = axs[i]
    ax.scatter(data[:, 0], data[:, 1], c=labels, cmap="viridis")
    ax.scatter(data[:, 0], data[:, 1], c=preds, cmap="viridis", alpha=0.5)
    
    #ax.scatter(X[labels == 0, 0], X[labels == 0, 1], s = 100, c = 'red', label = 'Cluster 1', alpha=0.1)
    #ax.scatter(X[labels == 1, 0], X[labels == 1, 1], s = 100, c = 'blue', label = 'Cluster 2', alpha=0.1)
    
    ax.set_title(f"k = {k}")
    ax.legend()

# save the plot
plt.tight_layout()
plt.savefig("kmeans_clusters_twogaussians42.png")
plt.show()
```
